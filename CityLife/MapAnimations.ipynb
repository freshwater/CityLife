```python
import IPython.display
from IPython.display import HTML, Javascript

import json
import folium

with open('data/coordinates_dates.json') as file:
    coordinates_dates = json.loads(file.read())
    
coordinates_dates = sorted(coordinates_dates, key=lambda cd: cd[1])

map_ = folium.Map(location=[39.833333, -98.583333],
			      zoom_start=4)

max_points = len(coordinates_dates);

data = {
    "map_id": map_.get_name(),
    "coordinates_dates": json.dumps(coordinates_dates[:max_points])
}

for coordinate, date in coordinates_dates[:max_points]:
    circle = folium.CircleMarker(
                    location=coordinate,
                    radius=2,
                    stroke=False,
                    fill_opacity=0,
                    fill_color='red')
    
    circle.add_to(map_)

js = Javascript("""
    function doTheThing() {
        var mapId = "%(map_id)s";
        var coordinates_dates = %(coordinates_dates)s;
        
        var colors = {};
        for (var cd of coordinates_dates) {
            var r = Math.round(Math.random()*255);
            var g = Math.round(Math.random()*255);
            var b = Math.round(Math.random()*255);
            
            colors[cd[1]] = [r, g, b];
        }

        // identify the HTML element corresponding to this map
        //  in case there are multiple maps on the page
        var iframes = document.getElementsByTagName('iframe');
        var doc = null;
        for (var iframe of iframes) {
            doc = iframe.contentWindow.document;

            if (doc.getElementById(mapId)) {
                break;
            }
        }

        // get all the markers. they are invisible initially
        var paths = doc.querySelectorAll('path');
        var count = paths.length;
        console.log(count);    

        // make them visible in sequence
        var intervalId = null;
        var index = 0;
        function tick() {
            if (index >= count - 1) {
                clearInterval(intervalId);
            }

            var rgb = colors[coordinates_dates[index][1]];
            var colorString = 'rgb(' + String(rgb) + ')';
            
            paths[index].setAttribute('fill', colorString);
            paths[index].setAttribute('fill-opacity', 0.5);
            index += 1;
        }

        intervalId = setInterval(tick, 10);

        console.log('intervalId', intervalId);
    }
    
    doTheThing();
""" % data)

IPython.display.display(map_)
IPython.display.display(js)
```
