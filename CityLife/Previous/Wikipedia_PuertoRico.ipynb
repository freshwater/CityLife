```python
soup = BeautifulSoup(http_get("https://en.wikipedia.org/wiki/List_of_communities_in_Puerto_Rico"))
```

```python
# list_ = soup.select('li > a[href^="/wiki/"]:not(.image):not(.internal)')

list_ = soup.select('a:first-child:not(.image):not(.external):not([href^="/wiki/Help:Category"])')

texts = [a.getText() for a in list_]
anchors = list_[texts.index('Z')+1:texts.index('^')-1]
page_titles = [a['href'][len('/wiki/'):] for a in anchors]
page_titles
```

```python
export_urls = [(title, "https://en.wikipedia.org/wiki/Special:Export/" + title) for title in page_titles]
soups = [(title, BeautifulSoup(http_get(url))) for title, url in export_urls]
```

---

```python
texts = [(title, soup.select('text')[0].getText()) for title, soup in soups]
```

```python
import pandas as pd
pd.set_option('display.max_rows', 1000)

websites = [[line for line in text.split("\n") if line.startswith('| web')]
            for title, text in texts]

df = pd.DataFrame(list(zip(page_titles, websites)))
df['x'] = df.apply(lambda x: 'http' in str(x[1]), axis=1)
df[df['x']]
```
